```python
x=num_df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
maxaccu=0
maxRS=0

for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    LR=LogisticRegression()
    LR.fit(x_train,y_train)
    pred=LR.predict(x_test)
    acc=accuracy_score(pred,y_test)
    if acc>maxaccu:
        maxaccu=acc
        maxRS=i

        
            
print('best accu score:-',maxaccu,'random state:-',maxRS)
        
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

```python
dtc_para={'criterion': ['gini','entropy'],'splitter':['best','random']}
```

```python
dtc=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
DTC=GridSearchCV(dtc,dtc_para,cv=5)
```

```python
DTC.fit(x,y)
```

```python
DTC.best_params_,DTC.best_score_
```

```python
DT_pred=DTC.best_estimator_.predict(x_test)
```

```python
print('accu score:-',accuracy_score(DT_pred,y_test))
print('conf matrix:-',confusion_matrix(DT_pred,y_test))
print('class report:-',classification_report(DT_pred,y_test))
```
